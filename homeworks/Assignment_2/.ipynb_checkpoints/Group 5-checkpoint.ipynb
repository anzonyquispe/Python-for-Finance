```python
#1. SHOW THE INDICES OF THE np.nan VALUES IN THE f_list. WE WANT TO SEE THIS OUTPUT: THE INDICES 0, 1, 4, 7 HAVE np.nan values. USING PRINT FUNCTION AND F-STRINGS TO INSERT THE INDICES VALUES
import numpy as np
f_list = [np.nan , np.nan, "Austria", "Germany", np.nan, "Pakistan", "np.nan", np.nan]
```

```python
#NOS PIDEN MOSTRAR LOS INDICES DE NP.NAN DE LA LISTA, PARA ELLO USAREMOS LA FUNCIÓN ENUMERATE QUE NOS DIRÁ LA POSICIÓN Y EL VALOR DE CADA ELEMENTO.
#PARA EVITAR ERRORES CON VALORES QUE NO SON NUMÉRICOS (COMO STRINGS), USAMOS ISINSTANCE PARA VERIFICAR QUE EL ELEMENTO SEA DE TIPO INT O FLOAT. LUEGO EJECUTAMOS NP.ISNAN() PARA COMPROBAR QUE ESE VALOR NUMÉRICO SEA NP.NAN.
#SI AMBAS CONDICIONES SE CUMPLEN, EL ÍNDICE DEL ELEMENTO SE GUARDA EN LA LISTA NAN_INDICES.
#FINALMENTE, USAMOS UN F-STRING PARA IMPRIMIR EL MENSAJE SOLICITADO Y LA FUNCIÓN JOIN (JUNTO CON MAP(STR)) PARA MOSTRAR LOS ÍNDICES SEPARADOS POR COMAS.
nan_indices = [
    i for i, x in enumerate(f_list)
    if isinstance(x, (int, float)) and np.isnan(x)]
print(f"The indices {', '.join(map(str, nan_indices))} have np.nan values. ")
```

```python
#1. Drop the duplicated blank spaces in str1. Do not use the regular expresions. Do not use the code shown in class. Explain your steps.Hint: Use split method and join function.
str1 = 'I am                            too                                                        old'
#PRIMERO USAMOS SPLIT() LO QUE DIVIDE EL STRING EN UNA LISTA DE PALABRAS
#CUANDO SE USA SPLIT() SIN ARGUMENTOS, SE ELIMINA AUTOMÁTICAMENTE LOS ESPACIOS EN BLANCO DUPLICADOS.
str1.split()
#USAMOS " ".join() PARA UNIR LOS ELEMENTOS DE LA LISTA EN UN SOLO STRING. CON JOIN INSERTAMOS UN SOLO ESPACIO ENTRE CADA PALABRA.
" ".join(['I', 'am', 'too', 'old'])
```

```python
#2. Get the number of letters in the string .Hint: Use lenfunction.
#PARA CONTAR SOLO LAS LETRAS, PRIMERO USAMOS EL RESULTADO DEL EJERCICIO ANTERIOR.
#USAMOS clean_str PARA LIMPIAR LOS ESPACIO DUPLICADOS DEL STRING
clean_str = " ".join(str1.split())
#COMO QUEREMOS SOLO LAS LETRAS, QUITAMOS LOS ESPACIOS:
only_letters = clean_str.replace(" ", "")
#PARA CONTAR LAS LETRAS USAMOS LA FUNCIÓN len()
num_letters = len(only_letters)
print(num_letters)
```

```python
#3. Get the number of blank spaces (all of them) in the string .Hint: Use lenfunction.
#SI CONTAMOS TODOS LOS CARACTERES Y LUEGO CONTAMOS EL STRING SIN ESPACIOS, LA DIFERENCIA SERÁN LOS ESPACIOS EN BLANCO
total_characters = len(str1)
#QUITAMOS TODOS LOS ESPACIOS Y LOS CONTAMOS CON LA FUNCION len()
no_spaces = str1.replace(" ", "")
characters_without_spaces = len(no_spaces)
#CALCULAMOS EL NÚMERO DE ESPACIOS
blank_spaces = total_characters - characters_without_spaces
print(blank_spaces)
```

```python
#4. Get the position of @ in each string in the emails list. Hint: Use mapfunction and find method.
#NOS PIDE LA POSICIÓN DEL @, PARA ELLO USAMOS EL MÉTODO FIND () QUE BUSCA UN CARÁCTER DENTRO DE UN STRING Y DEVUELVE LA POSICIÓN SI LO ENCUENTRA Y EL -1 SI NO EXISTE
#USAMOS MAP() PARA RECORRER CADA ELEMENTO DE LA LISTA emails.
#PARA CADA EMAIL, APLICAMOS FIND("@"), QUE DEVUELVE LA POSICIÓN DEL SÍMBOLO @.
#CONVERTIMOS EL RESULTADO EN UNA LISTA CON LIST() PARA PODER VISUALIZARLO.
positions = list(map(lambda x: x.find("@"), emails))
print(positions)
```

```python
#5. Identifies if exists .edu. in each string in the emails list. Get a list of Booleans. Hint: Use mapfunction and find method.
#USAMOS MAP() ESTO NOS PERMITE APLICAR UNA MISMA OPERACIÓN A CADA ELEMENTO DE UNA LISTA
#USAMOS LAMBDA PARA DEFINIR LA OPERACIÓN.
#EMAIL REPRESENTA CADA STRING DENTRO DE LA LISTA EMAILS
#.find(".edu.") BUSCA LA SUBCADENA .edu. DENTRO DEL EMAIL
#USAMOS FIND(), SI ENCUENTRA .edu. DEVUELVE LA POSICIÓN (0, 1, 2, etc) SI NO LA ENCUENTRA DEVUELVE -1
#FINALMENTE CONVERTIMOS EL RESULTADO EN UNA LISTA PARA OBTENER UNA LISTA DE VALORES BOOLEANOS
edu_exists = list(
    map(lambda email: email.find(".edu.") != -1, emails))
print(edu_exists)
```

```python
#6. Get all the strings before the first dot .in each string in the emails list. Identifies how many of them has @. Hint: Use mapfunction and find method.
#PRIMERO DEBEMOS OBTENER EL TEXTO ANTES DEL PRIMERO PUNTO PARA CADA EMAIL
#DE ESOS TEXTOS, DEBEMOS CONTAR CUANTOS CONTIENEN EL SÍMBOLO @
#OBTENEMOS LOS STRINGS ANTES DEL PRIMER PUNTO
#USAMOS MAP() PARA APLICAR ESTO A TODOS LOS EMAILS
before_dot = list(
    map(lambda email: email[:email.find(".")], emails))
#VERIFICAMOS CUÁLES CONTIENEN @ CON "@" IN X, TRUE SI EL STRING CONTIENE @, FALSE SI NO
#CONTAMOS CUANTOS CONTIENEN '@' O CONTAMOS CUÁNTOS TRUE HAY
count = sum(
    map(lambda x: "@" in x, before_dot))
print("Strings before first dot:", before_dot)
print("How many contain '@':", count)
```
