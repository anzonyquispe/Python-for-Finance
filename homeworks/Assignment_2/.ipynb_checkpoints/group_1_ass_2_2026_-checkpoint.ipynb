```python
#1 Tuple
tuple1 = ("Orange", [10, 20, 30], (5, 15, 25))
tuple2 = ([7, 8], [9, 1], [10, 7])
```

```python
#1.1 Tuple: The first item of the second item of tuple1 
print(tuple1[1][0])
```

```python
#1.2 Tuple:The last item of the tuple1
print(tuple1[2])==print(tuple1[-1])
```

```python
#1.3 Tuple: I wanted to change the value "orange" for "pink grapefruit" but It's impossible, because It's a tuple
try:
    tuple1[tuple1.index("Orange")] = "pink grapefruit"
except TypeError:
    print("No es posible")
#Tuples are immutable, so their values cannot be changed after creation
```

```python
#1.4 Tuple: This is the number of item by tuple1
len(tuple1)
```

```python
#1.5 Tuple: I use the funtion "map" for a sum the all of item by the tuple2
A = tuple(map(sum, tuple2))
print(sum(A))
```
