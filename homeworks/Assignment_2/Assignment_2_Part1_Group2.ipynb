```python
# =========================
# Kevin - Tuple 1.1 (Exercise 6)
# =========================
# Generate a new tuple (tuple3) with half values of tuple2
# Using map function, without loops

def half_list(lst):
    # map applies the lambda function to each element of the list
    return list(map(lambda x: x / 2, lst))

# Apply the function to each list inside tuple2
tuple3 = tuple(map(half_list, tuple2))

print("Original tuple2:", tuple2)
print("New tuple3 (half values):", tuple3)
```
